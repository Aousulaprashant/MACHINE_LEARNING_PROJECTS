Spam_Mail_Prediction_using_Machine_Learning
- importing libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
```

data collection and handling

```python
mail_data=pd.read_csv('/content/mail_data.csv')
mail_data.head()
```

```python
mail_data.isnull().sum()
mail_data.shape
```

LABEL ENCODING

```python
label_encode=LabelEncoder()
mail_data['Category']=label_encode.fit_transform(mail_data['Category'])
mail_data.head()
```

0 ->Ham mail            

1 ->Spam mail

```python
X=mail_data['Message']
Y=mail_data['Category']
```

SPLITING THE DATA INTO TRAINING AND TESTING DATASETS

```python
X_train, X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

EXTRACTING FEATURES

converting the text data into numarical values

```python
feature_extraction = TfidfVectorizer(min_df =1,stop_words='english',lowercase=True)

X_train_numarical = feature_extraction.fit_transform(X_train)
X_test_numarical = feature_extraction.transform(X_test)  # ensure we only transform the test set

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

creating model ,and feading the traind data to it

```python
model=LogisticRegression()

model.fit(X_train_numarical,Y_train)
```

```python
#predicting
#checkign the accury score

predict_train=model.predict(X_train_numarical)
accury_score=accuracy_score(predict_train,Y_train)
print(accury_score)
```

```python
predict_test=model.predict(X_test_numarical)
accury_score=accuracy_score(Y_test,predict_test)
print(accury_score)
```

```python
input=["Customer service annoncement. You have a New Years delivery waiting for you. Please call 07046744435 now to arrange delivery"]


input_data_features = feature_extraction.transform(input)


pridiction=model.predict(input_data_features)
print(pridiction)


if(pridiction == 0):
    print("ITS a Ham Mail")
else:
  print("its a spam mail")
```
